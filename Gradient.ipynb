```python
import matplotlib.pyplot as plt
import numpy as np

def compute(xo, lr, tolerance=1e-6, max_iter=1000):
    xnew = xo
    arr = []

    for i in range(max_iter):
        # Gradient descent update
        xprev = xnew  # Save the previous value for comparison
        xnew = xnew - lr * 2 * xnew  # Update rule

        # Store the new value in the array
        arr.append(xnew)

        # Early stopping condition
        if abs(xnew - xprev) < tolerance:
            print(f"Early stopping at iteration {i + 1}: x = {xnew}")
            break

    return arr, i + 1  # Return the array and number of iterations

# Parameters
xo = 1  # Initial x value
lr = 0.01  # Learning rate
tolerance = 1e-6  # Early stopping threshold

# Compute gradient descent
x, num_iters = compute(xo, lr, tolerance)

# Plotting the results
y = np.arange(num_iters)
plt.plot(y, x)
plt.xlabel('Iterations')
plt.ylabel('xnew')
plt.grid(True)
plt.title('Gradient Descent with Early Stopping')
plt.show()
```
